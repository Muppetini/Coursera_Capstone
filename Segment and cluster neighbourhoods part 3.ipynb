This is my notebook for segmenting and clustering neighbourhoods.

```python
import pandas as pd
import numpy as np
import urllib.request
from bs4 import BeautifulSoup
```

I am using urllib and Beautifulsoup to scrape and modify the information needed from Wikipedia.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
page = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
soup.title.string
```

The tables within the data are being found and then allocated to columns with new correct column names.

```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighbourhood']=C
df
```

This is to remove the 'Not assigned' values from the Borough column. Even though not shown in the dataframe, there is a \n referring to a new line that must be included.

```python
df_clean = df[df['Borough'] != 'Not assigned\n'].reset_index(drop=True)
df_clean.head()
```

```python
df_clean.shape
```

```python
df_clean['Postal Code'] = df_clean['Postal Code'].str.strip()
df_clean.head()
```

```python
!pip install geocoder
postal_data = pd.read_csv('http://cocl.us/Geospatial_data') 
postal_data.head()
```

```python
postal_data['Postal Code'] = postal_data['Postal Code'].str.strip()
```

```python
df_geo = pd.merge(df_clean, postal_data, on='Postal Code', how='left')
df_geo.head(12)
```

I am first stripping the Borough column for ease of selecting a borough. I am then focusing on North York as the borough to look at.

```python
df_geo['Borough'] = df_geo['Borough'].str.rstrip()
north_york = df_geo[df_geo['Borough'] == 'North York'].reset_index(drop=True)
north_york.head()
```

This is to import the required libraries.

I have found the latitue and longtitude for North Yorks in Toronto which I have used to create a map with markers.

```python
latitude = 43.76826
longitude = -79.41263

map_north_york = folium.Map(location=[latitude, longitude], zoom_start=11)

for lat, lng, label in zip(north_york['Latitude'], north_york['Longitude'], north_york['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_north_york)  
    
map_north_york
```
